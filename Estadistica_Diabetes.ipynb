```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datos = pd.read_csv('Book1.csv')
```

```python
datos.head(20)
```

```python
datos.info()
datos.shape
```

```python
datos = datos.drop(50, axis = 0)
```

```python
datos.info()
datos.shape
```

```python
datos
```

```python
Glucosa_ayunas = datos['Glucosa en ayunas (mg/dl)']
GA_normales = Glucosa_ayunas[Glucosa_ayunas<100]
GA_preDiabeticos = Glucosa_ayunas[(99<Glucosa_ayunas) & (Glucosa_ayunas<126)]
GA_diabeticos = Glucosa_ayunas[125<Glucosa_ayunas]
```

```python
mis_labels = ['Normales', 'Pre-diabeticos', 'Diabeticos']
plt.pie([GA_normales.size, GA_preDiabeticos.size, GA_diabeticos.size], labels = mis_labels)
plt.title('Glucosa en ayunas')
plt.show()
```

```python
datos['Glucosa en ayunas (mg/dl)'].hist(bins = 11, color = 'c',grid=True, edgecolor = "black", figsize=(10,7))
plt.xlabel('Glucosa (mg/dl)')
plt.ylabel('Frecuencia')
plt.title('Glucosa en ayunas', size=20)
```

```python
plt.boxplot(datos['Glucosa en ayunas (mg/dl)'])
plt.show()
```

```python
sns.boxplot(datos['Glucosa en ayunas (mg/dl)'])
```

```python
Hemoglobina_gluco = datos['Hemoglobina Glucosilada']
Hmg = Hemoglobina_gluco[pd.notna(Hemoglobina_gluco)]
Hmg_normales = Hmg[Hmg<5.7]
Hmg_preDiabeticos = Hmg[(Hmg>=5.7) & (Hmg<6.5)]
Hmg_diabeticos = Hmg[Hmg>=6.5]

```

```python
plt.pie([Hmg_normales.size, Hmg_preDiabeticos.size, Hmg_diabeticos.size], labels = mis_labels)
plt.title('Hemoglobina Glucosilada')
plt.show()
```

```python
datos['Hemoglobina Glucosilada'] = datos['Hemoglobina Glucosilada'].fillna(datos['Hemoglobina Glucosilada'].mean())
datos
```

```python
datos['Hemoglobina Glucosilada'].hist(bins = 11, color = 'c',grid=True, edgecolor = "black", figsize=(10,7))
plt.xlabel('Porcentaje %')
plt.ylabel('Frecuencia')
plt.title('Hemoglobina Glucosilada', size=20)
```

```python
sns.boxplot(datos['Hemoglobina Glucosilada'])
```

```python
plt.scatter(datos["Hemoglobina Glucosilada"], datos['Glucosa en ayunas (mg/dl)'])
plt.xlabel('Hemoglobina Glucosilada (%)')
plt.ylabel('Glucosa en ayunas (mg/dl)')
plt.title('Relacion Hemoglobina Glucosilada y Glucosa en ayunas', size=13)
```
